```python
import pandas as pd

df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")

df.head()
```

```python
# Convert churn column to numeric
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})

# Check churn distribution
df['Churn'].value_counts()
```

```python
# Overall churn rate
df['Churn'].mean()
```

```python
churn_rate = df['Churn'].mean()

print(f"Approximately {churn_rate*100:.2f}% of customers have churned, indicating potential retention issues.")
```

```python
# Churn by Gender
df.groupby('gender')['Churn'].mean()
```

```python
# Churn by Contract type
df.groupby('Contract')['Churn'].mean()
```

```python
# Churn by Internet Service
df.groupby('InternetService')['Churn'].mean()
```

```python
print("Business Insights:")

print("1. Customers with month-to-month contracts have significantly higher churn rates compared to long-term contracts.")

print("2. Certain customer segments (e.g., based on gender or service type) show higher churn risk.")

print("3. Customers with specific internet services (e.g., Fiber optic) tend to churn more, indicating possible service dissatisfaction.")
```

```python
import matplotlib.pyplot as plt

# Churn by Contract
df.groupby('Contract')['Churn'].mean().plot(kind='bar')

plt.title("Churn Rate by Contract Type")
plt.ylabel("Churn Rate")
plt.show()
```

```python
# Churn by Internet Service
df.groupby('InternetService')['Churn'].mean().plot(kind='bar')

plt.title("Churn Rate by Internet Service")
plt.ylabel("Churn Rate")
plt.show()
```

```python
print("Final Business Insights:")

print("• Approximately 26.5% of customers have churned, indicating a significant retention issue.")

print("• Customers with month-to-month contracts show the highest churn rates, suggesting lack of long-term commitment.")

print("• Customers using Fiber optic internet services tend to churn more, possibly due to pricing or service quality issues.")

print("• Long-term contract customers (1-year, 2-year) have significantly lower churn rates, highlighting the importance of customer retention strategies.")
```

```python
print("Business Recommendations:")

print("• Offer discounts or incentives to encourage customers to switch to long-term contracts.")

print("• Improve service quality or pricing strategies for high-churn segments like Fiber optic users.")

print("• Target at-risk customers (month-to-month users) with retention campaigns.")

print("• Introduce loyalty programs to reduce churn and improve customer retention.")
```
